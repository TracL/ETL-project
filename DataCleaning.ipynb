```python
redfin_file = 'Redfin_Scrape.csv'
trulia_file = 'Trulia_Scrape.csv'
hf_file = 'HomeFinder_Scrape.csv'

redfin_df = pd.read_csv(redfin_file)
trulia_df = pd.read_csv(trulia_file)
hf_df = pd.read_csv(hf_file)
```

```python
for i in hf_df.index:
    hf_df['City'][i] = hf_df['City'][i][:-4]
    hf_df['# of Beds'][i] = hf_df['# of Beds'][i][:hf_df['# of Beds'][i].find(' ')]
    hf_df['# of Baths'][i] = hf_df['# of Baths'][i][:hf_df['# of Baths'][i].find(' ')]
    hf_df['Price'][i] = hf_df['Price'][i].replace('$', '')
    hf_df['Price'][i] = hf_df['Price'][i].replace(',', '')
    
hf_df['# of Beds'] = hf_df['# of Beds'].astype(float)
hf_df['# of Baths'] = hf_df['# of Baths'].astype(float)
hf_df['Price'] = hf_df['Price'].astype(int)

hf_df
```

```python
for i in trulia_df.index:
    trulia_df['City'][i] = trulia_df['City'][i][trulia_df['City'][i].find(', ') + 2 : -4]
    trulia_df['City'][i] = trulia_df['City'][i][trulia_df['City'][i].find(', ') + 1 :]
    trulia_df['City'][i] = trulia_df['City'][i].lstrip()
    trulia_df['# of Beds'][i] = trulia_df['# of Beds'][i][:trulia_df['# of Beds'][i].find('b')]
    trulia_df['# of Baths'][i] = trulia_df['# of Baths'][i][:trulia_df['# of Baths'][i].find('b')]
    trulia_df['Price'][i] = trulia_df['Price'][i].replace('$', '')
    trulia_df['Price'][i] = trulia_df['Price'][i].replace(',', '')
    trulia_df['sqft'][i] = trulia_df['sqft'][i][:trulia_df['sqft'][i].find(' ')]
    trulia_df['sqft'][i] = trulia_df['sqft'][i].replace(',', '')
    
trulia_df['# of Beds'] = trulia_df['# of Beds'].astype(float)
trulia_df['# of Baths'] = trulia_df['# of Baths'].astype(float)
trulia_df['Price'] = trulia_df['Price'].astype(int)
trulia_df['sqft'] = trulia_df['sqft'].astype(int)

trulia_df
```

```python
for i in redfin_df.index:
    redfin_df['City'][i] = redfin_df['City'][i][: -4]
    redfin_df['# of Beds'][i] = redfin_df['# of Beds'][i][:redfin_df['# of Beds'][i].find(' ')]
    redfin_df['# of Baths'][i] = redfin_df['# of Baths'][i][:redfin_df['# of Baths'][i].find(' ')]
    redfin_df['Price'][i] = redfin_df['Price'][i].replace('$', '')
    redfin_df['Price'][i] = redfin_df['Price'][i].replace(',', '')
    redfin_df['sqft'][i] = redfin_df['sqft'][i][:redfin_df['sqft'][i].find(' ')]
    redfin_df['sqft'][i] = redfin_df['sqft'][i].replace(',', '')
    
redfin_df['# of Beds'] = redfin_df['# of Beds'].astype(float)
redfin_df['# of Baths'] = redfin_df['# of Baths'].astype(float)
redfin_df['Price'] = redfin_df['Price'].astype(int)
redfin_df['sqft'] = redfin_df['sqft'].astype(int)

redfin_df
```

```python
hf_df = hf_df.loc[hf_df['Price'] >= 10000]
hf_df = hf_df.loc[hf_df['# of Beds'] != 0]
hf_df = hf_df.loc[hf_df['# of Baths'] != 0]

hf_df
```

```python
hf_sq = [None] * 2954
tru_zip = [None] * 240

hf_df['sqft'] = hf_sq
trulia_df['Zip Code'] = tru_zip

trulia_df
```

```python
hf_df = hf_df[['Address', 'City', 'Zip Code', 'Price', 'sqft', '# of Beds', '# of Baths']]
trulia_df = trulia_df[['Address', 'City', 'Zip Code', 'Price', 'sqft', '# of Beds', '# of Baths']]
redfin_df = redfin_df[['Address', 'City', 'Zip Code', 'Price', 'sqft', '# of Beds', '# of Baths']]

redfin_df.columns, trulia_df.columns, hf_df.columns
```

```python
dfs = [hf_df, trulia_df, redfin_df]
all_dfs = pd.concat([df.squeeze() for df in dfs], ignore_index=True)

all_dfs
```

```python
all_dfs = all_dfs.drop(all_dfs.loc[all_dfs['Address'].duplicated() == True].index)
```

```python
all_dfs = all_dfs[all_dfs['Address'].notna()]
```
