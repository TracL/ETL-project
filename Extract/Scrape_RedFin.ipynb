```python
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
from splinter import Browser
import pandas as pd
```

### Setting up browser to scrape data using Splinter

```python
executable_path = {"executable_path": ChromeDriverManager().install()}
browser = Browser("chrome", **executable_path, headless=False)
```

### Scraping Current Listed Properties


```python
# Visit the following URL
url = "https://www.redfin.com/city/30844/TX/Frisco"
browser.visit(url)
```

```python
# Using BeautifulSoup to get html.parser
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Price of listed properties
price_list = []
price = soup.find_all('span', class_="homecardV2Price")
price_list = [p.text for p in price]
```

```python
print("Number of records in price list", len(price_list))
```

```python
# Numebr of Beds, Baths and Sq. Ft. of listed properties
beds_list = []
baths_list = []
sq_ft_list = []
stats = soup.find_all('div', class_="stats")

for each in stats:
    val = each.text
    if 'Bed' in val:
        beds_list.append(val.strip())
    elif 'Bath' in val:  
        baths_list.append(val.strip())
    elif 'Sq. Ft.' in val:
        sq_ft_list.append(val.strip())
```

```python
print("Number of records in beds list", len(beds_list))
print("Number of records in baths list", len(baths_list))
print("Number of records in sq. ft. list", len(sq_ft_list))
```

```python
# Zipcode of listed properties
zipcode_list = []
city_list = []
address_list = []

zipcode = soup.find_all('div', class_="link-and-anchor")
zipcode_list =  [z.text[-5:] for z in zipcode]
address_list = [z.text[:z.text.find(',')] for z in zipcode]
city_list = [z.text[z.text.find(',') + 2 : -6] for z in zipcode]

city_list
```

print("Number of records in zipcode list", len(zipcode_list))

### Creating a dataframe with sold real estate properties

```python
df = pd.DataFrame({"Price": price_list, "Address": address_list, "City": city_list, "Zip Code": zipcode_list, "# of Beds": beds_list, "# of Baths": baths_list, "sqft": sq_ft_list})
df
```

```python
df.to_csv("../CSVs/Redfin_Scrape.csv")
```
